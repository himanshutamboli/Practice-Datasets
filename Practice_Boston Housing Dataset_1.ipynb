### Exploring abbreviation in the dataset

- CRIM - per capita crime rate by town
- ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
- INDUS - proportion of non-retail business acres per town.
- CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
- NOX - nitric oxides concentration (parts per 10 million)
- RM - average number of rooms per dwelling
- AGE - proportion of owner-occupied units built prior to 1940
- DIS - weighted distances to five Boston employment centres
- RAD - index of accessibility to radial highways
- TAX - full-value property-tax rate per 10,000 dollars.
- PTRATIO - pupil-teacher ratio by town
- B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
- LSTAT - % lower status of the population
- MEDV - Median value of owner-occupied homes in 1000's dollars

### Performing Univariate Analysis to the dataset

#### What is the purpose of Univariate Analysis?

- In the dataset, we will do analysis on only one/single column
- Using UA, we can determine the distribution of datapoints
- Based on distribution of datapoints, we can make assumptions

```python
features=[feature for feature in bh.columns if bh[feature].dtype !='0']
print('We got {} numerical features: {}'.format(len(features), features))
```

```python
plt.figure(figsize=(15,25))
plt.suptitle('Univariate Analysis of Numerical features',fontsize=20,fontweight='bold',alpha=1,y=0)

for i in range(0,len (features)):
    plt.subplot(5,3,i+1)
    sb.histplot(x=bh[features[i]],color='r',line_kws={'color':'blue','linestyle':'dashed'},kde=True)
    plt.xlabel(features[i])
    plt.tight_layout()
```

### Performing Multivariate Analysis to the dataset

#### What is the purpose of Multivariate Analysis?

- In the dataset, we will compare two or more features/columns
- Using MA, we can determine which feature is dependent/independent
- Based on it, we can check the relation between them e.g. correlation

```python
bh[bh.columns].corr()
```

#### Observation from the above Statistical description, we can determine:

- If the correlation is -ve -- correlation is decreasing with the target variable
- If the correlation is +ve -- correlation is increasing with the target variable

```python
plt.figure(figsize=(15,15))
sb.heatmap(bh.corr(),annot=True)
plt.show()
```

#### Observation from the above heatmap, below are the considerations:

##### Target variable id 'medv' is Median value of owner-occupied homes in 1000$
- Heatmap is the best graphical representation of correlation between the variables/features
- Darket the color, lesser the correlation
- Lighter the color, higher teh correlation
- medv has high +ve correlation with the rm (RM - average number of rooms per dwelling)
- medv has high -ve correlation with lstat (LSTAT - % lower status of the population)

### Performing Data and Information visualization on the dataset

```python
plt.subplots(figsize=(15,10))
sb.histplot(bh.medv,bins=20,kde=True,color='blue')
plt.title('MEDV Distribution',fontsize=20,pad=20)
plt.xlabel('Price in $1000',fontsize=15)
plt.ylabel('Count',fontsize=15)
```

#### Observation from the above graph, below are the considerations:

- We can see the graph is slightly right skewed
- The max number of people who owns the house has income between **_15-25k $_**
- There are some people who has income below **_10k_** and **_40k_**

```python
# Used weight=bold as font

plt.subplots(figsize=(15,10))
sb.histplot(bh.rm)
plt.title('Average number of rooms per dwelling',weight='bold',fontsize=20,pad=20)
plt.xlabel('No. of rooms',weight='bold',fontsize=15)
plt.ylabel('Count',weight='bold',fontsize=15)
```

#### Observation from the above graph, below is the considerations:

- The most average room per dwelling is in between 5.5 to 7
