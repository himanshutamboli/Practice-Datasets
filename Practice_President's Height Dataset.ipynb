# First way of EDA on President's Height dataset

```python
import numpy as np
import pandas as pd
import warnings # for ignoring any warnings
warnings.filterwarnings('ignore')
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
sb.set(style='white', color_codes=True)
```

```python
ph=pd.read_csv('president_heights.csv')
ph
```

```python
heights=ph['height']
print('Mean of height:',heights.mean())
print('Standard Deviation of height:',heights.std())
print('Minimum of height:',heights.min())
print('Maximum of height:',heights.max())
```

## np.percentile function used to find the percentile in NumPy

```python
print('25th percentile:', np.percentile(heights,25))
print('50th percentile:', np.percentile(heights,50))
print('75th percentile:', np.percentile(heights,75))
```

```python
plt.hist(heights)
plt.title('Height distribution of US Presidents')
plt.xlabel('Heights (cm)')
plt.ylabel('Number')
```
