```python
# What is Subplot? How to make 2 graph in same area?
# What is (1,2,1)? Is there 1_Row; 2_Column and again 1_Index
```

```python
x=np.arange(1,11)
```

```python
y1=x*2
y2=x*3
```

```python
y1
```

```python
y2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linewidth=3)
plt.show()
```

```python
# Pie Chart
# We can see the distribution between different Categorical value and we see them in percentage
```

```python
fruit=['Apple','Mango','Guava','Grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
fruit=['Apple','Mango','Guava','Grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','yellow','green','blue'])
plt.show()
```

```python
# Doughnut Chart
```

```python
fruit=['Apple','Mango','Guava','Grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit,radius=1.5,colors=['red','yellow','green','blue'],autopct='%0.1f%%')
plt.pie([1],radius=1,colors='w')
plt.show()
```
